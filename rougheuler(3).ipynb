```python
import numpy as np
from matplotlib import pyplot as plt

t_i = 0
y_i = 1
t_f = 10
h = 0.05

t_s = [t_i]
FE_s = [y_i]
BE_s = [y_i]

t_e = np.linspace(t_i,t_f,1000)
y_e = []
for t in t_e:
    y_e.append(np.exp(-10*t))
plt.plot(t_e, y_e)
plt.title('exact')
plt.xlabel('t')
plt.ylabel('y')
plt.show()

#FE approx
t = t_i
while t<t_f:
    yn = FE_s[-1]
    ynp1 = (1.0-2.0*h*t)*yn
    FE_s.append(ynp1)
    t = t+h
    t_s.append(t)
    
plt.plot(t_e, y_e)
plt.plot(t_s, FE_s, '-.')
plt.title('FE approx')
plt.xlabel('t')
plt.ylabel('y')
plt.show()

t_s = [t_i]
t = t_i
while t<t_f:
    yn = BE_s[-1]
    ynp1 = yn/(1.0+2.0*h*t)
    BE_s.append(ynp1)
    t = t+h
    t_s.append(t)
    
plt.plot(t_e, y_e)
plt.plot(t_s, BE_s, '-.')
plt.title('BE approx')
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
